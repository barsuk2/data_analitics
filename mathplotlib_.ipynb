```python
import matplotlib.pyplot as plt
import numpy as np
from pylab import rcParams
x = np.linspace(0,10,50)
y = x
rcParams['figure.figsize'] = 12,9

plt.plot(x,y, label= 'Линейня функция')
plt.rc_context({'xtick.color':'white', 'ytick.color':'white'})
plt.title('Линейная зависимость' ,color = 'w', fontsize = 'xx-large', loc = 'left') # заголовок
plt.xlabel('Ось абсцисс', color = 'w', fontsize = 14, fontstyle = 'italic') # ось абсцисс
plt.ylabel('Ось ординат', color = 'w', fontsize = 14, fontstyle = 'italic') # ось ординат
plt.text(7,2,'Текстовое примечание', color = 'r', fontsize = 14,)
plt.legend()
plt.grid() # установить сетку

```

## Размещение графиков отдельно
